## Completing the Cyber Kill Chain

![Cyber Kill Chain](./images/cyber-kill-chain.jpeg)

Process
1. Create a map of the Cyber Kill Chain.
2. Identify where on the Kill Chain the incident was detected and fill this out.
3. Work you way down the Kill Chain to step 7, filling out anything found in the logs. Add MITRE ATT&CK mappings where relevant.
4. Once you reach the end (or the attack was stopped at a specific point), go back up the Kill Chain to find where it started from.
5. Use indicators gathered to pivot to new data points and investigate if anything else was affected.
